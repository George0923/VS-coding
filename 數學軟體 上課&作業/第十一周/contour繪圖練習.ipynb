```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.linspace(-3*np.pi, 3*np.pi)
y = np.linspace(0, 9*np.pi)
X, Y = np.meshgrid(x,y)
Z = np.sin(X) + np.cos(Y)
```

```python
plt.contourf(X,Y,Z, cmap=plt.cm.hot, alpha=0.9);
plt.contour(X,Y,Z, cmap='tab20b');
```
