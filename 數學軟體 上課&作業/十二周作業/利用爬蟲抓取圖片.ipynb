```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import requests
from bs4 import BeautifulSoup
import shutil
```

```python
res = requests.get('https://www.niusnews.com/=P3pkt814')
soup = BeautifulSoup(res.text)
for img in soup.select('.img'):
    fname = img['scr'].split('/')[-1]
    res2 = requests.get(img['scr'], stream=True)
    f = open(fname, 'wb')
    shutil.copyfileobj(res2.raw, f)
    f.close()
    del res2
```
