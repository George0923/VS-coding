## 刪去多餘表格

```python
currency = df[0].ix[:,0:5]
```

```python
currency.columns = [u'幣別', u'現金匯率-本行買入', u'現金匯率-本行賣出', u'即期匯率-本行買入', u'即期匯率-本行賣出']
```

```python
currency[u'幣別'] = currency[u'幣別'].str.extract('\((\w+)\)')
```
