# 詞雲生成(投資學課本)

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import jieba
```

```python
path = open(r'C:\Users\USER\Desktop\大學\大三\三下\投資學\投資學 課本.pdf',encoding='big5',errors='ignore').read()
```

```python
wordlist = jieba.cut(path, cut_all = True)
words = " ".join(wordlist)
wordcloud = WordCloud().generate(words)
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
