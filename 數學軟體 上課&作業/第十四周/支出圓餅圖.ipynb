```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
food = int(input("本周的食物支出金額為:"))
drink = int(input("本周的飲料支出金額為:"))
clothes = int(input("本周的服飾支出金額為:"))
fun = int(input("本周的娛樂支出金額為"))
labels = 'food', 'drink', 'clothes', 'fun'
sizes = [food, drink, clothes, fun]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```
