# 利用爬蟲功能抓取匯率

```python
import pandas as pd
```

```python
df = pd.read_html('https://rate.bot.com.tw/xrt?Lang=zh-TW')
# 讀入網站
```

```python
df[0]
```

```python
currency = df[0].ix[:,0:5]
currency.columns = [u'幣別', u'現金匯率-本行買入', u'現金匯率-本行賣出', u'即期匯率-本行買入', u'即期匯率-本行賣出']
currency[u'幣別'] = currency[u'幣別'].str.extract('\((\w+)\)')
# 整理資料及格式
```
