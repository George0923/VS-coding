```python
os.chdir(f'{lion_file_path}/lion') ## into lion file

Y = os.listdir() ## get years
print(Y)
for i in range(len(Y)):
    os.chdir(f'{lion_file_path}/lion/{Y[i]}') ## into year file
    
    M = os.listdir() ## get months
    print(Y[i], M)
    for j in range(len(M)):
        os.chdir(f'{lion_file_path}/lion/{Y[i]}/{M[j]}') ## into month file
        
        ALL = os.listdir() ##get all file
        print(Y[i], M[j], ALL)
        for k in range(len(ALL)):
            df_tmp = pd.read_csv(ALL[k], index_col = False).rename(columns = {"Unnamed: 0":"code"})
            df_tmp.loc[:,"date"] = ALL[k][:-4]
            df = pd.concat( [df,df_tmp] )
```
